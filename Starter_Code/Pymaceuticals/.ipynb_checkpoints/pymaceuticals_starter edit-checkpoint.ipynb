```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata,study_results,on='Mouse ID')
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.

#Remove all uncessary columns 
check_df = combined_data[['Mouse ID','Timepoint']]

#Create a new dataframe with the duplicate data only
duplicate_df = check_df[check_df.duplicated(keep=False)]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = combined_data.loc[combined_data['Mouse ID']=='g989']
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID
clean_data = combined_data.loc[combined_data['Mouse ID']!= 'g989']
clean_data
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
clean_data.columns
```

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

#Remove uncessary columns 
tumor_df = clean_data[['Drug Regimen','Tumor Volume (mm3)']]

#Calculates mean of tumors for each drug
drug_mean = tumor_df.groupby(by='Drug Regimen').mean(numeric_only =True)
#Renaming column
drug_mean = drug_mean.rename(columns = {'Tumor Volume (mm3)':'Mean Tumor Volume'})

#Calculates Median for tumors for each drug
drug_median = tumor_df.groupby(by='Drug Regimen').median(numeric_only =True)
#Renaming column
drug_median = drug_median.rename(columns = {'Tumor Volume (mm3)':'Median Tumor Volume'})

#Calculate variance of tumors for each drug
drug_var = tumor_df.groupby(by='Drug Regimen').var(numeric_only =True)
#Renaming colum
drug_var = drug_var.rename(columns = {'Tumor Volume (mm3)':'Tumor Volume Variance'})

#Calculate standard deviation for each drug 
drug_std = tumor_df.groupby(by='Drug Regimen').std(numeric_only =True)
#Renaming colum
drug_std = drug_std.rename(columns = {'Tumor Volume (mm3)':'Tumor Volume STD'})

#Calculate SEM of tumor volume for each drug
drug_sem = tumor_df.groupby(by='Drug Regimen').sem(numeric_only =True)
#Renaming colum
drug_sem = drug_sem.rename(columns = {'Tumor Volume (mm3)':'Tumor Volume SEM'})



# Assemble the resulting series into a single summary DataFrame.
drug_summary = pd.concat([drug_mean, drug_median, drug_var, drug_std, drug_sem],axis=1)
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
drug_summary_2 = tumor_df.groupby(by='Drug Regimen').agg([np.mean,np.median,np.var,np.std,st.sem])
drug_summary_2
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
#Defining x_axis points 
x_axis = np.arange(len(clean_data['Drug Regimen'].unique()))

#A list of each unique drug name
drug_list = [drug for drug in clean_data['Drug Regimen'].unique()]

#Grab y-axis values, the total count for mouse ID and timepoint for each drug 
y_axis = [clean_data.loc[clean_data['Drug Regimen'] == drug ]['Mouse ID'].count() for drug in drug_list]
```

```python
for drug in drug_list:
    print(f"{clean_data.loc[clean_data['Drug Regimen'] == drug ]['Mouse ID'].count()} {drug}")
```

```python
#Plotting bar graph
bar_plot_df = pd.DataFrame(data = {'x_axis':x_axis,'y_axis':y_axis})
bar_plot_df = bar_plot_df.sort_values('y_axis',ascending=False)
plt.bar(x_axis,bar_plot_df['y_axis'])


#Setting x axis labels using tick locations 
tick_locations = [value for value in x_axis]
#Getting drug name in order of highest counts of rows 
labels = clean_data.groupby(by = 'Drug Regimen').count().sort_values('Mouse ID',ascending=False).index
plt.xticks(tick_locations,labels,rotation='vertical')

#Setting x axis and y axis labels
plt.xlabel('Drug Regimen')
plt.ylabel('# of Observed Mouse timepoints')

#Displaying the bar graph
plt.tight_layout()
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
bar_panda = (clean_data[['Drug Regimen','Mouse ID']])
bar_panda.groupby(by = 'Drug Regimen').count().sort_values('Mouse ID',ascending=False).plot(kind='bar',\
                                                                                            ylabel='# of Observed Mouse timepoints',\
                                                                                           legend=False)
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
clean_data.groupby('Sex')['Mouse ID'].count().plot(kind='pie',\
                                                   subplots=True,\
                                                   autopct = '%1.1f%%',\
                                                  ylabel = 'Sex',\
                                                  startangle = 180,\
                                                  colors = ['Orange','Blue'])
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
#Creating a label list for all the unique genders in the dataframe
labels = clean_data['Sex'].unique()
#Colors to be used for the plot
colors = ['Orange','Blue']
#Creating a list for count of each gender 
sizes = clean_data.groupby('Sex')['Mouse ID'].count().values
#List to specify where to cut for the pie graph
explode = [0,0.1]

#Ploting the pie graph using pyplot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,\
       autopct = '%1.1f%%',\
        startangle = 180)
#Changing ylabel 
plt.ylabel('Sex')
plt.show()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
#Removing uncessary drugs from dataframe by isolating each drug name
df1 = clean_data.loc[clean_data['Drug Regimen']=='Capomulin']
df2 = clean_data.loc[clean_data['Drug Regimen']=='Ramicane']
df3 = clean_data.loc[clean_data['Drug Regimen']=='Infubionol']
df4 = clean_data.loc[clean_data['Drug Regimen']=='Ceftamin']

#Merge all isolated drug name dataframe into one
final_tumor = pd.concat([df1,df2,df3,df4])

# Start by getting the last (greatest) timepoint for each mouse
Mouse_id = final_tumor['Mouse ID'].unique()
last_timepoint = [final_tumor.loc[final_tumor['Mouse ID']==Mouse][['Mouse ID','Timepoint']].max() for Mouse in Mouse_id]

#Turning last_timepoint into a dataframe
last_timepoint = pd.DataFrame(last_timepoint)
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_vol = pd.merge(final_tumor,last_timepoint,on=['Mouse ID','Timepoint'],how='right')
```
