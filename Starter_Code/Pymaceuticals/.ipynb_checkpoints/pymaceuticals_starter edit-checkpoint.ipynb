```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np 

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata,study_results,on='Mouse ID')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
```

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

#Remove uncessary columns 
tumor_df = clean_data[['Drug Regimen','Tumor Volume (mm3)']]

#Calculates mean of tumors for each drug
drug_mean = tumor_df.groupby(by='Drug Regimen').mean(numeric_only =True)
#Renaming column
drug_mean = drug_mean.rename(columns = {'Tumor Volume (mm3)':'Mean Tumor Volume'})

#Calculates Median for tumors for each drug
drug_median = tumor_df.groupby(by='Drug Regimen').median(numeric_only =True)
#Renaming column
drug_median = drug_median.rename(columns = {'Tumor Volume (mm3)':'Median Tumor Volume'})

#Calculate variance of tumors for each drug
drug_var = tumor_df.groupby(by='Drug Regimen').var(numeric_only =True)
#Renaming colum
drug_var = drug_var.rename(columns = {'Tumor Volume (mm3)':'Tumor Volume Variance'})

#Calculate standard deviation for each drug 
drug_std = tumor_df.groupby(by='Drug Regimen').std(numeric_only =True)
#Renaming colum
drug_std = drug_std.rename(columns = {'Tumor Volume (mm3)':'Tumor Volume STD'})

#Calculate SEM of tumor volume for each drug
drug_sem = tumor_df.groupby(by='Drug Regimen').sem(numeric_only =True)
#Renaming colum
drug_sem = drug_sem.rename(columns = {'Tumor Volume (mm3)':'Tumor Volume SEM'})



# Assemble the resulting series into a single summary DataFrame.
drug_summary = pd.concat([drug_mean, drug_median, drug_var, drug_std, drug_sem],axis=1)
```

```python
drug_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line

#Using groupby and aggregate to retrieve the mean, median, variance, standard deviation and SEM for tumor volume for 
#each drug regimen
drug_summary_2 = tumor_df.groupby(by='Drug Regimen').agg([np.mean,np.median,np.var,np.std,st.sem])
#Display the summary table 
drug_summary_2
```

## Quartiles, Outliers and Boxplots

```python
# A DatFrame that has the last timepoint for each mouse ID is created using groupby.
timepoint_mouse = clean_data.loc[clean_data.groupby(by='Mouse ID')['Timepoint'].idxmax()]

# The index of the DataFrame is reset. 
timepoint_mouse = timepoint_mouse.reset_index(drop=True)

# Retrieve the maximum timepoint for each mouse.
maximum_timepoint = timepoint_mouse[['Mouse ID','Timepoint']]
```

```python
# The four treatment groups, Capomulin, Ramicane, Infubinol, and Ceftamin, are put in a list. (3 points)
treatment = ['Capomulin','Ramicane','Infubinol','Ceftamin']

# An empty list is created to fill with tumor volume data. (3 points)
tumor_vol_data = []
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

#Removing uncessary drugs from dataframe by isolating each drug name
df1 = clean_data.loc[clean_data['Drug Regimen']=='Capomulin']
df2 = clean_data.loc[clean_data['Drug Regimen']=='Ramicane']
df3 = clean_data.loc[clean_data['Drug Regimen']=='Infubinol']
df4 = clean_data.loc[clean_data['Drug Regimen']=='Ceftamin']

#Merge all isolated drug name dataframe into one
final_tumor = pd.concat([df1,df2,df3,df4])
```

```python
# Getting the greatest timepoint for each mouse 
Mouse_id = final_tumor['Mouse ID'].unique()
last_timepoint = [final_tumor.loc[final_tumor['Mouse ID']==Mouse][['Mouse ID','Timepoint']].max() for Mouse in Mouse_id]

#Turning last_timepoint into a dataframe
greatest_timepoint = pd.DataFrame(last_timepoint)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint for each mouse
final_tumor_vol = pd.merge(final_tumor,greatest_timepoint,on=['Mouse ID','Timepoint'],how='right')
```

```python
# Display the final tumor volume dataframe for each mouse
final_tumor_vol
```

```python
# A for loop is used to display the interquartile range (IQR) and the outliers for each treatment group (10 points)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

#For loop to check and print the potential outliers for each drug regimen treatment
for drug in treatment:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes 
    drug_tumor = final_tumor_vol.loc[final_tumor_vol['Drug Regimen'] == drug]
    
    #and add them to tumor_vol_data list
    tumor_vol_data.append(drug_tumor['Tumor Volume (mm3)'])
    
    #Determine outliers using upper and lower bounds
    #Defining quartiles and IQR 
    quartiles = drug_tumor['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

    #Creating upper and lower bound for outliers 
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_vol.loc[(final_tumor_vol['Drug Regimen'] == drug) & ((final_tumor_vol['Tumor Volume (mm3)'] < lower_bound) |\
                                  (final_tumor_vol['Tumor Volume (mm3)']> upper_bound))]
    #Printing drug name and it potential outliers with it index from final_tumor_vol
    print(f"{drug}")
    print(f"potential outliers:{outliers['Tumor Volume (mm3)']}")
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
#Create one figure to store all the boxplots
fig1, ax1 = plt.subplots()
#Plot tumor_vol_data as a boxplot
plt.boxplot(tumor_vol_data,flierprops = dict(markerfacecolor='red'))
#Change x_axis labels to match drug name
xticks = [value +1  for value in np.arange(len(treatment))]
plt.xticks(xticks ,treatment)
#display figure with all the boxplots
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

#Defining slope, intercept, rvalue, pvalue and standard deviation
(slope, intercept, rvalue,pvalue,stdev) = st.linregress(x_axis, y_axis)

#Defining regress_value 
regress_value = x_axis * slope + intercept

#Defining line equation
line_equation = f"y = {round(slope,2)}x + {round(intercept,2)}"

#Printing line equation and correlation coefficient 
print(f"The line equation for Weight vs Tumor volume is {line_equation}")
print(f"The correlation coefficient for Weight vs Tumor volume is {round(rvalue,2)}")
```

```python
#Replotting the scatter plot 
plt.scatter(x_axis,y_axis, marker="o", facecolors="red", edgecolors="black",
            s=x_axis, alpha=0.75)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
#Adding in the linear equation to the graph
plt.plot(x_axis,regress_value,"--",label= line_equation)
#Adding linear equation label to the graph
plt.annotate(line_equation, (18,37),fontsize= 15,color='blue')
#Tight layout and displaying the final graph
plt.tight_layout()
plt.show()
```
